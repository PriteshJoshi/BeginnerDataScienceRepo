```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris #import iris flower dataset
```

```python
iris=load_iris()
```

```python
df= pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
df.head() # print first 5 records
```

```python
#summary of data, min, max, mean, range, std, varience, percentile
summary=df.describe()
summary=summary.transpose()

print(summary)

# calculating Varience
varience=summary.loc['sepal length (cm)']['std']**2
print('\n\nvarience for spetal length: {:.3f} '.format(varience))

varience=summary.loc['sepal width (cm)']['std']**2
print('\n\nvarience for spetal width: {:.3f} '.format(varience))

varience=summary.loc['petal length (cm)']['std']**2
print('\n\nvarience for petal lenth: {:.3f} '.format(varience))

varience=summary.loc['petal width (cm)']['std']**2
print('\n\nvarience for petal width: {:.3f} '.format(varience))
```

```python
#histogram for sepal length  
plt.hist(df['sepal length (cm)'],bins=15)
plt.title('Sepal Length')
plt.xlabel('sepal lentgh ')
plt.ylabel('count')
plt.show()
```

```python
#histogram for sepal width 
plt.hist(df['sepal width (cm)'],bins=15)
plt.title('Sepal Width')
plt.xlabel('sepal width ')
plt.ylabel('count')
plt.show()
```

```python
#histogram for petal length  
plt.hist(df['petal length (cm)'],bins=15)
plt.title('petal Length')
plt.xlabel('petal lentgh')
plt.ylabel('count')
plt.show()
```

```python
#histogram for petal width
plt.hist(df['petal width (cm)'],bins=15)
plt.title('petal width')
plt.xlabel('petal width ')
plt.ylabel('count')
plt.show()
```
